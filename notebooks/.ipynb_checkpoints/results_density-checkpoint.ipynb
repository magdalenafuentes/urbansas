```python
def compute_stats(res_dict: dict, index: dict, recall_levels) -> dict:
    """
    Compute statistics
    Args:
        res_dict: results dictionary, on entry per uid with at least tp, fp, fn
        index: index dictionary
        recall_levels: Recall levels to compute precision at
    Returns:
        dictionary of statistics
    """

    stats_dict = {}
    for subset in ['global', 'mavd', 'tau']:

        def condition(file_dict):
            if subset == 'global':
                return True
            if subset == 'mavd':
                return file_dict['city'] == 'montevideo'
            if subset == 'tau':
                return file_dict['city'] != 'montevideo'

        uids = [uid for uid, file_dict in index.items() if condition(file_dict)]
        
        stats_frame = {}
        for frame_type in ['active_frames', 'inactive_frames', 'all_frames']:

            if len(uids):
                # Sum tp, fp, fn across all uids, for different values of dynamic_range
                tp = np.sum(np.stack([np.array([res[frame_type]['tp'] for res in res_dict[uid]]) for uid in uids]), axis=0)
                fp = np.sum(np.stack([np.array([res[frame_type]['fp'] for res in res_dict[uid]]) for uid in uids]), axis=0)
                fn = np.sum(np.stack([np.array([res[frame_type]['fn'] for res in res_dict[uid]]) for uid in uids]), axis=0)

                precision = np.where(tp > 0, tp / (tp + fp), 0)
                recall = np.where(tp > 0, tp / (tp + fn), 0)
                stats_frame[frame_type] = { f'precision@{r:.2f}': precision[np.argmin(np.abs(recall - r))]
                                for r in recall_levels
                                }
                              
                
        stats_dict[subset] = stats_frame

    return stats_dict

def load_index(index_filepath):
    """
    Currently, one of:
        - "TAU2021_final.json": index used for initial training/predictions (per Julia MIR project)
        - "TAU2021_after_max_box_filtering.json": new index used when only the largest 
            bounding box per class/frame is considered
    
    """
    INDEX_PATH = Path(os.path.join('', index_filepath)) # PASS IN YOUR INDEX FILEPATH HERE
    index = pd.read_json(INDEX_PATH).transpose()
    with INDEX_PATH.open() as f:
        index = json.load(f)

    df_index = pd.DataFrame.from_dict(index,orient='index')
    df_index.head()
    print("Length of index: ", len(index))
    return index

def format_json(dicpaths, model_name, recall_levels=[0.1, 0.2, 0.3, 0.4]):
    res_dict = {}
    for path in dicpaths:
        res_dict.update(json.load(open(path)))
    index = load_index("../index/Urbansas.json")
    stats = {model_name:compute_stats(res_dict, index, recall_levels)}
    return stats
```
