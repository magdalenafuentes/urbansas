```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import json
import os
import librosa
from librosa import display
from itertools import cycle
import IPython
from moviepy.editor import AudioFileClip,VideoFileClip,CompositeVideoClip, VideoClip
from moviepy.video.io.bindings import mplfig_to_npimage
from matplotlib.patches import Rectangle
import glob
import h5py
import seaborn as sns
```

```python
RESULTS_PATH = '../results_icassp'
```

```python
all_res = pd.DataFrame({})
for results_path in glob.glob(os.path.join(RESULTS_PATH, '*.csv')):
    res = pd.read_csv(results_path).drop('Unnamed: 0', axis=1)
    res['model'] = res['name']
    all_res = pd.concat([all_res, res])
```

```python
all_res
```

```python
all_res.model.unique()
```

```python
score = 'iou'
melted_scores = pd.DataFrame({})
for model in res.model.unique():
    num_regions = 5#int(model.split('/')[-2].split(',')[1].split('-')[-1])
#     in_dur = int(model.split('/')[-2].split(',')[3].split('-')[-1])
    _res = all_res#[(res.model==model)&(res.score==score)].copy()
    _res = _res.groupby(['score', 'tau', 'frame_type', 'uid', 'model', 'model_type']).mean().reset_index()
    _res = pd.melt(_res, value_vars=['bus', 'car', 'motorbike', 'truck'], 
                   id_vars=['uid', 'frame_type', 'tau', 'model', 'model_type'], 
                   var_name='class', value_name=score)
    _res['num_regions'] = num_regions
#     _res['in_dur'] = in_dur
    melted_scores = pd.concat([melted_scores, _res])
melted_scores = melted_scores.sort_values('num_regions')
melted_scores
```

```python
ax = sns.catplot(x='tau', y=score, hue='model', data=melted_scores, col='frame_type', row='model_type',
                 kind="bar", height=6, aspect=1)
# xticks = np.arange(0, 8, 0.05)
# plt.xticks(xticks, xticks)
```
