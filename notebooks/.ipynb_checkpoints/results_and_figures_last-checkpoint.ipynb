```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import json
import os
import librosa
from librosa import display
from itertools import cycle
import IPython
from moviepy.editor import AudioFileClip,VideoFileClip,CompositeVideoClip, VideoClip
from moviepy.video.io.bindings import mplfig_to_npimage
from matplotlib.patches import Rectangle
import glob
import h5py
import seaborn as sns
```

```python
RESULTS_PATH = '../results/'
```

```python
all_res = pd.DataFrame({})
for path in glob.glob(os.path.join(RESULTS_PATH,'*.csv')):
    print(path)
    res = pd.read_csv(path).drop('Unnamed: 0', axis=1)
    res['model_type'] = path.split('_')[-1].split('.')[0]
    all_res = pd.concat([all_res, res])
```

```python
all_res.model.unique()
```

```python
score = 'iou'
melted_scores = pd.DataFrame({})
for model in all_res.model.unique():
    num_regions = 5#int(model.split(',')[1].split('-')[-1])
    _res = all_res[(all_res.model==model)&(all_res.score==score)].copy()
    _res = _res.groupby(['score', 'tau', 'frame_type', 'uid', 'model_type']).mean().reset_index()
    _res = pd.melt(_res, value_vars=['bus', 'car', 'motorbike', 'truck'], 
                   id_vars=['uid', 'frame_type', 'tau', 'model_type'], 
                   var_name='class', value_name=score)
    _res['num_regions'] = num_regions
    _res['model'] = model
#     _res['loss'] = model.split('/')[-2].split('_')[0]
#     _res['stratification'] = model.split('/')[-2].split('_')[-1]
    _res['loss'] = model.split('/')[3].split('__')[1].split('-')[-1]
    _res['loss'] = _res['loss'].map({'BinaryCrossentropy':'BCE', 'WeightedBinaryCrossentropy':'WBCE'})
    _res['stratification'] = model.split('/')[3].split('__')[2].split('-')[-1]
    _res['stratification'] = _res['stratification'].map({'location_id':'location', 'city':'city'})
    melted_scores = pd.concat([melted_scores, _res])
melted_scores = melted_scores.sort_values('num_regions')
melted_scores = melted_scores.dropna()
melted_scores
```
