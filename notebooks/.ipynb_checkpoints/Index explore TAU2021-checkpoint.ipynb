```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append('../')

import project_paths
import project_params

%matplotlib notebook
```

# Load TAU2021 index

```python
index_path = project_paths.index_folder.joinpath('TAU2021.json')
with index_path.open() as f:
    index = json.load(f)
```

```python
df_index = pd.DataFrame.from_dict(index,orient='index')
df_index
```

# Analyze azimuth distribution

```python
ds_azimuth = df_index['events'].map(lambda events: [ list(az for az in event['azimuth']) for event in events if 'azimuth' in event  ])
ds_azimuth = ds_azimuth.map(lambda x: np.concatenate(x) if len(x) else np.zeros(0))
```

```python
all_azimuth = np.concatenate(ds_azimuth.to_numpy())
```

```python
num_regions = project_params.num_regions
fov = project_params.fov
region_angle = fov/num_regions

plt.figure(figsize=(9,4))
plt.hist(all_azimuth,bins=np.linspace(-fov/2,fov/2,num_regions+1))
plt.grid()
plt.title(f'Azimuth distribution, {num_regions} regions, {region_angle:.0f} deg per region')
plt.xlabel('Azimuth [deg]')
plt.ylabel('Count')
plt.tight_layout()
```
