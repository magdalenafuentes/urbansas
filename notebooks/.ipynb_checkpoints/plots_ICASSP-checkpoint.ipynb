```python
import os
import sys
import glob
import numpy as np
import pandas as pd
from tqdm import tqdm
import soundfile as sf
import seaborn as sns
import matplotlib.pyplot as plt
from collections import namedtuple 

URBANSAS_DIR = '/Users/mf3734/sound_datasets/urbansas/'
```

```python
audio_csv = os.path.join(URBANSAS_DIR, 'annotations', 'audio_annotations.csv')
video_csv = os.path.join(URBANSAS_DIR, 'annotations', 'video_annotations.csv')
```

```python
audio = pd.read_csv(audio_csv)
video = pd.read_csv(video_csv)
audio
```

```python
len(audio.filename.unique()), len(video.filename.unique())
```

```python
audio.groupby('non_identifiable_vehicle_sound').count().reset_index()/len(audio)
```

```python
video.groupby(['city', 'filename']).count()
```

```python
df_vehicles = video.track_id.groupby([video.frame_id, video.label, video.filename]).nunique().to_frame().reset_index()
df_vehicles = df_vehicles[df_vehicles.label!='-1']
df_vehicles['frames'] = 1
df_vehicles['frames'].loc[df_vehicles.frame_id==-1] = 20  # adjust number of frames
df_vehicles['minutes'] = 0.5/60 * df_vehicles['frames']
df_vehicles
```

```python
colors = ["#52b788", "#007b7f", "#118ab2", "#3d405b"]
sns.set_palette(sns.color_palette(colors))
# import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
sns.barplot(x="label", y="frames", data=df_vehicles.groupby('label').sum().reset_index().sort_values('frames', ascending=False), 
            palette=sns.color_palette(['#007b7f', '#007b7f','#007b7f','#007b7f']), )
plt.ylabel('Ocurrences (frames)', fontsize=12)
plt.xlabel('', fontsize=12)
plt.xticks([0,1, 2, 3], ['car', 'bus', 'motorbike', 'truck'], fontsize=12)
plt.savefig('figs/unique_vehicles.png', dpi=300, bbox='tight')
```

```python
stats = pd.DataFrame()
v = 'non_identifiable_vehicle_sound'
_df = audio.groupby([v, 'filename']).count().reset_index().loc[:,[v, 'filename']]
# tau = video[video.subset=='tau'].filename.unique()       
# _df = _df[_df.filename.isin(tau)]
_df = _df.groupby(v).count()/len(_df)
_df = _df.reset_index()
stats = pd.concat([stats, 
                   pd.DataFrame({'metric':v, 
                                 'percentage': _df[_df[v]==1].filename})])

v = 'night'
_df = video.groupby([v, 'filename']).count().reset_index().loc[:,[v, 'filename']]
_df = _df.groupby(v).count()/len(_df)
_df = _df.reset_index()
stats = pd.concat([stats, 
                   pd.DataFrame({'metric':v, 
                                 'percentage': _df[_df[v]==1].filename})])


stats
```

```python
sns.barplot(x="metric", y="percentage", data=stats*100, 
            palette=sns.color_palette(['#007b7f', '#007b7f']))
plt.ylabel('Percentage', fontsize=12)
plt.xlabel('', fontsize=12)
plt.xticks([0,1], ['Non-identifiable vehicle sounds', 'Night'], fontsize=12)
plt.savefig('figs/flags.png', dpi=300, bbox='tight')
```
