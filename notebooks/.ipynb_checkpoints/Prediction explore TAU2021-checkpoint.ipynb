```python
import os
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import h5py
from pathlib import Path
import librosa
import librosa.display
from moviepy.editor import AudioFileClip,VideoFileClip,CompositeVideoClip, VideoClip
from moviepy.video.io.bindings import mplfig_to_npimage
from itertools import cycle

import sys
sys.path.append('../')

import project_paths
import project_params

%matplotlib notebook
```

# Load TAU2021 index

```python
ROOT_DIR = os.path.abspath(os.path.join(os.path.abspath(''), '..'))
DATAPATH = os.path.abspath(os.path.join(os.path.abspath(''), '..', '..', 'urban_sight-and-sound', 'dataset', 'processed'))
```

```python
index_path = project_paths.index_folder.joinpath('TAU20211.json')
with index_path.open() as f:
    index = json.load(f)
```

```python
df_index = pd.DataFrame.from_dict(index,orient='index')
df_index
```

# Load predictions

```python
dataset = 'TAU2021'
config_name = 'tau2021_f0_scratch'
fold = 0

with project_paths.checkpoints_folder.joinpath(config_name,'params.json').open('r') as f:
    train_params = json.load(f)
classes = train_params['classes']

pred_path = list(project_paths.predictions_folder.joinpath(config_name).glob(f'{dataset}-f{fold}-*.h5'))[0]
pred = {}
with h5py.File(pred_path,'r' ) as f:
    for uid in f.keys():
        pred[uid] = {
            'time': np.array(f[uid]['time']),
            'pred': np.array(f[uid]['pred']),
        }
```

```python
pred.keys()
```

# Visualize prediction

```python
%matplotlib inline
```

```python
video_root = Path(os.path.join(DATAPATH, 'video'))
audio_root = Path(DATAPATH)

# for uid in ['1260' 1279 '190', '200', '205', '207', '209', '211', '219', '220', '224', '227', '228', '235', '238', '241', '242', '331', '333', '337', '339', '342', '353', '378']:

uid = '395'

file_index = index[uid]
file_pred = pred[uid]

print(f'Visualizing prediction for uid: {uid} {file_index["path"]}')

audio_path = audio_root.joinpath(file_index['path'])
video_path = video_root.joinpath(Path(file_index['path']).stem + '.mp4')

# Load audio and display ground truth
file_audio, file_sr = librosa.load(audio_path,sr=None,mono=True)

fig, axs = plt.subplots(2,1,figsize=(18, 12),sharex='all')

# Plot wave and spectrogram
plt.sca(axs[0])
librosa.display.waveshow(file_audio,sr=file_sr)
plt.grid()

plt.sca(axs[1])
file_audio_stft = librosa.stft(file_audio)
file_audio_specgram = librosa.amplitude_to_db(np.abs(file_audio_stft), ref=np.max)
librosa.display.specshow(file_audio_specgram,sr=file_sr,y_axis='mel',x_axis='time')

# Add ground truth
bar_height = 0.4
bar_bottom_it = cycle(2**np.arange(8,15,0.5))

plt.sca(axs[1])
for event in file_index['events']:
    # print(event['time'])
    if 'time' in event:
        if not event['time']: 
            continue
        start = event['time'][0]
        end = event['time'][-1] + project_params.labels_period
    else:
        start = event['start']
        end = event['end']
    print(event)
    ctr = (start+end)/2
    width = end-start
    bottom = next(bar_bottom_it)
    height = bar_height*bottom
    plt.bar(x=ctr,width=width,bottom=bottom,height=height,alpha=0.5)
    plt.text(start+0.05,bottom+height/2,f'({event["source"][:1].upper()}) {event["label"]}',color='w',va='center')

plt.show()
```

```python
angles = np.linspace(-train_params['fov']/2,train_params['fov']/2,train_params['num_regions'])

# Load video and display a static frame with predictions
file_video = VideoFileClip(str(video_path), target_resolution=(480,720))

fig, axs = plt.subplots(1,3,figsize=(20,5))

def make_frame(frame_time:float):
    
    pred_time_idx = np.argmin(np.abs(frame_time-file_pred['time']))


    axs[0].clear()
    axs[0].imshow(file_video.get_frame(frame_time))

    axs[1].clear()
    axs[1].imshow(np.fliplr(file_pred['pred'][pred_time_idx]),vmin=0,vmax=0.8,cmap='magma')
    axs[1].axis('auto')
    axs[1].set_yticks(np.arange(len(classes)))
    axs[1].set_yticklabels(classes)
    axs[1].set_xticks(np.arange(len(angles)))
    axs[1].set_xticklabels([ f'{a:.0f}' for a in angles[::-1] ])
    axs[1].set_xlabel('Azimuth [deg]')

        
    file_gt = np.zeros((15, 5))
    for d in file_index['events']:
        if 'time' in d and d['confirmed']:
            for t, az in zip(d['time'], d['azimuth']):
                if abs(t - frame_time) < 0.5:
                    # print(az//1, project_params.fov/2, az//15)
                    i_az = int((az + project_params.fov/2) / project_params.fov * 15)
                    i_lbl = [-1, 'bus', 'car', 'motorbike', 'truck'].index(d['label'])
                    file_gt[i_az, i_lbl] = 1
                    # print(file_gt[:,2], [int((az + project_params.fov/2)//15)])

    # axs[2].clear()
    # axs[2].imshow(np.fliplr(file_gt.T),vmin=0,vmax=1,cmap='magma')
    # axs[2].axis('auto')
    # axs[2].set_yticks(np.arange(len(classes)))
    # axs[2].set_yticklabels(classes)
    # axs[2].set_xticks(np.arange(len(angles)))
    # axs[2].set_xticklabels([ f'{a:.0f}' for a in angles[::-1] ])
    # axs[2].set_xlabel('Azimuth [deg]')
    
    file_gt = np.zeros((15, 5))
    for d in file_index['events']:
        if 'time' in d and d['confirmed']:
            for t, az in zip(d['time'], d['azimuth']):
                if abs(t - frame_time) < 0.5:
                    # print(az//1, project_params.fov/2, az//15)
                    i_az = int((az + project_params.fov/2) / project_params.fov * 15)
                    i_lbl = [-1, 'bus', 'car', 'motorbike', 'truck'].index(d['label'])
                    file_gt[i_az, i_lbl] = 1
                    # print(file_gt[:,2], [int((az + project_params.fov/2)//15)])

    axs[2].clear()
    axs[2].imshow(np.fliplr(file_gt.T),vmin=0,vmax=1,cmap='magma')
    axs[2].axis('auto')
    axs[2].set_yticks(np.arange(len(classes)))
    axs[2].set_yticklabels(classes)
    axs[2].set_xticks(np.arange(len(angles)))
    axs[2].set_xticklabels([ f'{a:.0f}' for a in angles[::-1] ])
    axs[2].set_xlabel('Azimuth [deg]')

    # plt.tight_layout()

    return mplfig_to_npimage(fig)

# Render and display
file_audio_clip = AudioFileClip(str(audio_path))
file_video_pred = VideoClip(make_frame, duration=file_video.duration).set_audio(file_audio_clip)
file_video_pred.ipython_display(fps=2, autoplay=True)
```
