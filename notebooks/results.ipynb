```python
import os
import pandas as pd
```

```python
RESULTSPATH = '../results'

csvs = [ 
    'results_urbansas_f0_point_sources.csv',
    'results_urbansas_f1_point_sources.csv',
    'results_urbansas_f2_point_sources.csv',
    'results_urbansas_f3_point_sources.csv',
    'results_urbansas_f4_point_sources.csv',  
    'results_urbansas_f0_box_sources.csv',
    'results_urbansas_f1_box_sources.csv',
    'results_urbansas_f2_box_sources.csv',
    'results_urbansas_f3_box_sources.csv',
    'results_urbansas_f4_box_sources.csv',  
]
```

```python
df = pd.DataFrame({})
for csv in csvs:
    df = pd.concat([df, pd.read_csv(os.path.join(RESULTSPATH, csv))]).drop('Unnamed: 0', axis=1)
    
_df = df.groupby(['score', 'tau', 'frame_type', 'uid', 'model_type']).mean().reset_index()
_df = pd.melt(_df, value_vars=['bus', 'car', 'motorbike', 'truck'], 
               id_vars=['uid', 'frame_type', 'tau', 'model_type'], 
               var_name='class', value_name='iou')
```
